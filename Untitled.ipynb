```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

pd.pandas.set_option('display.max_columns', None)
```

```python
df =pd.read_csv('house_price.csv')
df.head()
```

```python
df.shape
```

# In Data Analysis we will analyze to find out the below stuff

1. Missing Values
2. All the Numerical Variable
3. Distributionof Numerical Varibles
4. Categorical Catiable
5. Cardinality of Categorical Varibales
6. Outliers
7. Relationship between independent and depenedent Features(Sale Price)

```python
features_with_na = [features for features in df.columns if df[features].isnull().sum() > 1]
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean() , 4), ' % Missing Values')
```

```python
for feature in features_with_na:
    data  = df.copy()

    data[feature] = np.where(data[feature].isnull(), 1, 0)

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_columns = df.select_dtypes(include = ['int64', 'float64']).columns
print("Number of Numerical Columns:", len(numerical_columns))
df[numerical_columns].head()
```

```python
categorical_columns = df.select_dtypes(include = ['object']).columns
print( 'Numerical Columns : ', len(categorical_columns))
```

```python
year_feature = [feature for feature in numerical_columns if 'Yr' in feature or 'Year' in feature]  
year_feature
```

```python
 for feature in year_feature:
     print(feature, df[feature].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
plt.show()
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        
        data = df.copy()
        
        # Create new feature
        data[feature] = data['YrSold'] - data[feature]
        
        # Scatter plot
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```
