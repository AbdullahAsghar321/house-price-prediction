```python
features_with_na = [features for features in df.columns if df[features].isnull().sum() > 1]
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean() , 4), ' % Missing Values')
```

```python
for feature in features_with_na:
    data  = df.copy()

    data[feature] = np.where(data[feature].isnull(), 1, 0)

    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_columns = [feature for feature in df.columns if df[feature].dtypes != 'O']
print('Numerical Columns : ', len(numerical_columns))
```

```python
year_features = [feature for feature in numerical_columns if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_features:
    print(feature, df[feature].unique())
```

```python
## let analyze the data temporal datetime variable
## we will check whether there is a relation between year the house is hold or sale price

df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
plt.show()
```

```python
## Hera we will compare the difference between All years feature with YrSold

for feature in year_features:
    if feature != 'YrSold':
        data = df.copy()
        data[feature] = data['YrSold']-data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Saleprice')
        plt.show()
```

```python
discrete_feature = [feature for feature in numerical_columns if len(df[feature].unique()) < 25 and feature not in year_features+['Id']]
print(discrete_feature)
```

```python
for feature in discrete_feature:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature = [feature for feature in numerical_columns if len(df[feature].unique()) > 25 and feature not in discrete_feature+year_features+['Id']]
print('Continuous Feature Count {}'.format(len(continuous_feature)))
```

```python
for feature in continuous_feature:
    data = df.copy()
    data[feature].hist(bins= 25)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
## we will be using logarthmetic transformation

for feature in continuous_feature:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
    
```

## Outliers

```python
for feature in continuous_feature:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
